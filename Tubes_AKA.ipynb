```python
class Novel:
    def __init__(self, name, pages, year, rating):
        self.name = name
        self.pages = pages
        self.year = year
        self.rating = rating


def init_novels():
    return [
        Novel("Langit Senja", 245, 2005, 8.2),
        Novel("Pelangi di Matamu", 310, 1999, 8.7),
        Novel("Jejak Tak Terlihat", 400, 2012, 7.8),
        Novel("Sang Bayangan", 360, 2018, 9.1),
        Novel("Hati yang Tersenyum", 220, 1994, 8.5),
        Novel("Rindu Senja", 290, 2003, 8.3),
        Novel("Embun Pagi", 280, 1990, 7.9),
        Novel("Ombak Kehidupan", 450, 2017, 8.6),
        Novel("Mimpi di Ujung Jalan", 320, 1998, 7.4),
        Novel("Cahaya Remang", 375, 2015, 8.9),
        Novel("Gelap Gulita", 290, 2002, 7.6),
        Novel("Matahari Malam", 410, 2008, 8.8),
        Novel("Tetes Air Hujan", 230, 2019, 8.1),
        Novel("Dunia tanpa Bintang", 300, 1996, 7.3),
        Novel("Langkah Terakhir", 360, 2010, 9.0),
        Novel("Angin Timur", 290, 2013, 8.2),
        Novel("Suara Hati", 320, 2021, 8.7),
        Novel("Pintu Masa Lalu", 275, 2001, 7.8),
        Novel("Rintik Hujan", 410, 2007, 8.5),
        Novel("Lara Hati", 345, 1993, 7.9),
        Novel("Bayangan Surga", 450, 2016, 9.1),
        Novel("Jalan Setapak", 260, 2009, 7.4),
        Novel("Cermin Waktu", 280, 2020, 8.6),
        Novel("Pesona Hati", 310, 2014, 8.3),
        Novel("Hati yang Lelah", 330, 2006, 7.7),
        Novel("Kidung Malam", 400, 1997, 8.4),
        Novel("Langit yang Hilang", 390, 2018, 8.8),
        Novel("Sisi Lain Dunia", 275, 1991, 7.5),
        Novel("Percikan Asa", 350, 2004, 8.1),
        Novel("Lukisan Jiwa", 310, 2012, 7.9),
        Novel("Malam di Ufuk Timur", 280, 1992, 7.6),
        Novel("Cahaya dalam Gelap", 440, 2017, 8.9),
        Novel("Denting Waktu", 350, 2000, 8.7),
        Novel("Kabut di Atas Bukit", 315, 2013, 8.4),
        Novel("Dunia Tak Berujung", 300, 2005, 7.8),
        Novel("Laut yang Tenang", 400, 2022, 9.2),
        Novel("Pohon Harapan", 250, 2009, 8.3),
        Novel("Jejak di Pasir", 310, 1999, 8.2),
        Novel("Hembusan Angin", 270, 2011, 7.7),
        Novel("Di Bawah Bayangan", 320, 2020, 8.9),
        Novel("Kisah Senja", 380, 2003, 7.4),
        Novel("Perjalanan Jiwa", 330, 2006, 8.5),
        Novel("Bunga di Padang Gersang", 345, 2018, 8.1),
        Novel("Puncak Rindu", 400, 2015, 7.6),
        Novel("Angin Laut", 390, 1994, 8.8),
        Novel("Awan Mendung", 290, 2002, 7.3),
        Novel("Mimpi Merah", 410, 2008, 9.1),
        Novel("Malam Tanpa Bulan", 320, 2021, 8.6),
        Novel("Pintu Harapan", 350, 2007, 8.2),
        Novel("Sinar Pagi", 260, 1993, 7.8),
        Novel("Bayangan di Balik Dinding", 280, 2016, 8.4),
        Novel("Lembah Asa", 300, 1995, 7.5),
        Novel("Cahaya Tak Terlihat", 440, 2010, 8.7),
        Novel("Suara di Keheningan", 350, 2019, 9.0),
        Novel("Langkah Sepi", 315, 2013, 8.3),
        Novel("Pelangi Kehidupan", 300, 1998, 7.9),
        Novel("Sang Penantang", 280, 2004, 8.6),
        Novel("Dunia Baru", 400, 2022, 9.2),
        Novel("Aroma Hujan", 250, 2005, 8.1),
        Novel("Bintang yang Hilang", 310, 2012, 8.4),
        Novel("Kidung Angin", 275, 1997, 7.6),
        Novel("Bolpenku", 200, 1980, 8.0),
        Novel("Hujan", 211, 2014, 8.6),
        Novel("Lautan Sepi", 350, 1995, 7.5),
        Novel("Bintang Merah", 430, 2020, 9.2),
        Novel("Dunia Fantasi", 280, 2010, 6.8),
        Novel("Senja di Ujung Jalan", 310, 2018, 8.9),
        Novel("Mimpi di Balik Awan", 250, 2005, 7.0),
        Novel("Kabut Hitam", 500, 1988, 9.0),
        Novel("Rembulan Malam", 350, 2016, 8.2),
        Novel("Petualangan Rahasia", 390, 1992, 6.5),
        Novel("Langit Biru", 410, 2021, 9.5),
        Novel("Jejak Waktu", 230, 1997, 7.8),
        Novel("Mawar Berduri", 260, 1985, 5.4),
        Novel("Tarian Ombak", 300, 2001, 7.6),
        Novel("Cahaya dalam Gelap", 290, 2019, 9.1),
        Novel("Mentari Pagi", 220, 2012, 8.3),
        Novel("Hilang dalam Hening", 400, 1990, 6.9),
        Novel("Perjalanan Rahasia", 350, 2015, 8.7),
        Novel("Kota Bayangan", 450, 2008, 7.4),
        Novel("Tepian Harapan", 310, 2003, 8.1),
        Novel("Surga yang Hilang", 270, 1998, 6.7),
        Novel("Di Balik Bukit", 290, 2017, 9.3),
        Novel("Gelora Jiwa", 240, 1994, 7.2),
        Novel("Pelangi Hati", 360, 1982, 8.4),
        Novel("Rindu Tak Bertepi", 340, 2013, 8.0),
        Novel("Melodi Cinta", 320, 1999, 7.9),
        Novel("Kenangan Manis", 200, 1989, 6.6),
        Novel("Kasih Tak Sampai", 410, 2022, 9.4),
        Novel("Gadis Desa", 290, 2006, 7.3),
        Novel("Badai di Puncak", 420, 1993, 6.8),
        Novel("Hati yang Terluka", 280, 1987, 5.9),
        Novel("Asmara di Ujung Jalan", 330, 2000, 7.7),
        Novel("Lukisan Senja", 310, 2018, 9.6),
        Novel("Bulan di Atas Langit", 370, 2011, 8.2),
        Novel("Cinta Tak Berbalas", 410, 2009, 6.3),
        Novel("Bening Embun Pagi", 220, 1986, 7.0),
        Novel("Rahasia Bintang", 360, 2023, 9.8),
        Novel("Horizon Merah", 290, 1996, 8.8),
        Novel("Lentera Kecil", 320, 1983, 7.1),
        Novel("Bayangan Malam", 250, 2020, 8.5),
        Novel("Jejak Bayangmu", 400, 1992, 6.4),
        Novel("Perjalanan Asa", 300, 2014, 7.5),
        Novel("Jingga Senja", 290, 1991, 6.2),
        Novel("Warna-Warni Dunia", 340, 2007, 8.0),
        Novel("Cermin Hati", 410, 2019, 9.7),
        Novel("Langkah Kecil", 280, 2016, 8.3),
        Novel("Ombak Putih", 390, 2004, 7.8),
        Novel("Kilau Permata", 430, 2021, 8.9),
        Novel("Bunga Mawar", 310, 1981, 6.0),
        Novel("Hati Biru", 200, 2010, 7.2),
        Novel("Bintang Malam", 320, 1982, 7.8),
        Novel("Langit Kelabu", 240, 1991, 8.3),
        Novel("Pelangi di Timur", 370, 2004, 6.7),
        Novel("Hujan Deras", 280, 2007, 7.5),
        Novel("Mentari Senja", 290, 2012, 8.2),
        Novel("Jejak Bayangan", 410, 2015, 9.1),
        Novel("Mimpi Indah", 310, 1985, 6.4),
        Novel("Matahari Baru", 350, 1990, 8.6),
        Novel("Rahasia Alam", 300, 1998, 7.9),
        Novel("Cahaya Surga", 410, 2020, 9.4),
        Novel("Lembah Biru", 260, 1986, 5.9),
        Novel("Awan Mendung", 390, 1993, 6.3),
        Novel("Puncak Harapan", 280, 1999, 8.0),
        Novel("Rindu Abadi", 320, 2002, 8.7),
        Novel("Kabut Pekat", 360, 2014, 7.4),
        Novel("Dunia Gelap", 310, 2008, 6.1),
        Novel("Mawar Biru", 410, 2016, 9.0),
        Novel("Titik Balik", 290, 1997, 8.4),
        Novel("Badai Mimpi", 330, 2019, 7.3),
        Novel("Harapan Baru", 280, 2021, 9.7),
        Novel("Jejak Petualang", 300, 1980, 8.5),
        Novel("Langkah Kecil", 390, 1995, 7.1),
        Novel("Gema Alam", 270, 2003, 6.9),
        Novel("Kota Lama", 350, 1989, 7.2),
        Novel("Senandung Rindu", 220, 1992, 7.6),
        Novel("Hening Senja", 290, 2000, 8.1),
        Novel("Suara Ombak", 420, 2010, 9.8),
        Novel("Ruang Gelap", 300, 1994, 6.5),
        Novel("Bunga Musim Gugur", 410, 2023, 9.6),
        Novel("Kisah Awan", 200, 1984, 6.0),
        Novel("Langit Terbuka", 330, 2018, 7.8),
        Novel("Harapan Hati", 290, 1991, 8.3),
        Novel("Pelangi Pagi", 400, 2005, 9.2),
        Novel("Rindu Senja", 280, 1987, 7.4),
        Novel("Dunia Pelangi", 320, 2011, 6.8),
        Novel("Cahaya Malam", 350, 2014, 8.9),
        Novel("Rahasia Hati", 260, 2019, 7.0),
        Novel("Lukisan Masa Lalu", 410, 1990, 9.3),
        Novel("Langit Malam", 300, 2008, 7.7),
        Novel("Goresan Senja", 400, 2001, 6.9),
        Novel("Kepingan Hati", 310, 2013, 8.5),
        Novel("Badai dan Cahaya", 290, 2016, 7.2),
        Novel("Jejak Rahasia", 280, 2004, 8.0),
        Novel("Puncak Gunung", 410, 2017, 9.4),
        Novel("Hujan Pagi", 320, 1995, 7.3),
        Novel("Langit Tak Berujung", 280, 2020, 9.1),
        Novel("Cahaya di Timur", 350, 2006, 8.7),
        Novel("Angin Senja", 360, 2012, 6.8),
        Novel("Rindu Mendalam", 310, 1998, 7.9),
        Novel("Bunga di Jalan", 200, 2015, 8.4),
        Novel("Langkah Terakhir", 330, 1993, 7.6),
        Novel("Mentari Tengah Malam", 300, 1983, 6.7),
        Novel("Jejak Langit", 400, 2018, 9.0),
        Novel("Ombak dan Karang", 290, 1982, 7.0),
        Novel("Bayang Senja", 420, 1994, 8.2),
        Novel("Rindu dan Duka", 310, 2009, 7.8),
        Novel("Langit Berbintang", 410, 2000, 9.5),
        Novel("Badai Senyap", 280, 2021, 6.2),
        Novel("Kabut dan Rindu", 370, 2013, 8.6),
        Novel("Jingga Malam", 360, 1986, 7.5),
        Novel("Bunga Sakura", 200, 1988, 6.1),
        Novel("Lukisan Pelangi", 290, 1996, 8.1),
        Novel("Harapan Pagi", 320, 2010, 7.7),
        Novel("Langkah Baru", 310, 2002, 7.3),
        Novel("Jejak Masa Depan", 280, 2019, 9.4),
        Novel("Cinta dan Ombak", 400, 2005, 6.6),
        Novel("Bulan dan Bintang", 290, 2022, 9.7),
        Novel("Hati yang Lembut", 350, 1981, 6.0),
        Novel("Langit di Timur", 300, 1999, 7.8),
        Novel("Badai Berakhir", 420, 2017, 8.3),
        Novel("Cahaya Harapan", 400, 1980, 7.1),
        Novel("Rindu yang Hilang", 280, 2003, 6.5),
        Novel("Jingga dan Senja", 390, 2014, 8.0),
        Novel("Puncak Keheningan", 410, 2007, 8.9),
        Novel("Rahasia Malam", 320, 2023, 9.2),
        Novel("Hati Merindu", 200, 1989, 6.8),
        Novel("Bayang Pelangi", 290, 1994, 7.5),
        Novel("Jejak Mimpi", 350, 2020, 9.1),
        Novel("Langit Kelam", 300, 1984, 7.6),
        Novel("Senandung Jiwa", 310, 2018, 8.2),
        Novel("Harapan Senja", 360, 2016, 8.7),
        Novel("Dunia yang Baru", 420, 1992, 7.9),
        Novel("Dunia yang Baru", 420, 1992, 7.9),
        Novel("Bolpenku", 200, 1980, 8.0),
        Novel("Perang yang Tidak Pernah Selesai", 350, 2005, 8.3),
        Novel("Jejak Sejarah", 500, 2010, 9.0),
        Novel("Kehidupan di Zaman Kuno", 600, 2000, 7.5),
        Novel("Langit Merah di Atas Dunia", 310, 1998, 8.0),
        Novel("Menembus Waktu", 450, 2015, 8.6),
        Novel("Sejarah Abad Ke-20", 420, 2003, 7.8),
        Novel("Peradaban yang Hilang", 380, 2008, 8.1),
        Novel("Pahlawan yang Terlupakan", 330, 1990, 8.5),
        Novel("Di Balik Layar Sejarah", 400, 2012, 7.7),
        Novel("Raja dan Rakyat", 470, 1995, 8.2),
        Novel("Jejak Kemenangan", 510, 2001, 8.4),
        Novel("Semesta yang Tak Terungkap", 560, 2018, 7.6),
        Novel("Perjalanan Seorang Pejuang", 530, 2007, 8.0),
        Novel("Di Tengah Perang", 490, 1999, 7.9),
        Novel("Perang Dunia Pertama", 550, 2002, 8.3),
        Novel("Revolusi yang Mengubah Dunia", 430, 1993, 8.7),
        Novel("Peradaban yang Berdiri Kembali", 560, 2010, 8.5),
        Novel("Kehidupan di Era Klasik", 450, 2004, 8.1),
        Novel("Sejarah Pemberontakan", 520, 2006, 8.8),
        Novel("Sisa-Sisa Peradaban", 470, 2009, 7.7),
        Novel("Zaman Emas", 500, 1996, 8.2),
        Novel("Warna Perjuangan", 410, 2000, 7.8),
        Novel("Legenda Para Pahlawan", 460, 2001, 8.0),
        Novel("Kota yang Hilang", 550, 1997, 8.5),
        Novel("Penyusup Waktu", 380, 2014, 8.1),
        Novel("Menjadi Sejarah", 600, 2003, 8.6),
        Novel("Langkah Ke Abad Baru", 490, 2016, 8.3),
        Novel("Menyusuri Masa Lalu", 460, 2005, 8.2),
        Novel("Sisi Gelap Sejarah", 520, 2017, 7.9),
        Novel("Bersama Sejarah", 480, 1994, 8.4),
        Novel("Tantangan Abad Ini", 500, 2008, 7.8),
        Novel("Jejak Para Sejarawan", 530, 2006, 8.3),
        Novel("Menapak Sejarah", 410, 2007, 8.5),
        Novel("Perubahan Zaman", 460, 2009, 8.0),
        Novel("Pahlawan di Bayangan", 520, 2012, 8.7),
        Novel("Zaman Kemegahan", 500, 2010, 8.2),
        Novel("Legenda Sejarah", 450, 1991, 8.4),
        Novel("Bumi yang Menyaksikan", 550, 2002, 7.7),
        Novel("Menghadapi Masa Depan", 490, 2011, 8.3),
        Novel("Dunia yang Terlupakan", 600, 1997, 8.5),
        Novel("Tangan yang Membangun Sejarah", 520, 2003, 8.1),
        Novel("Jejak Peradaban", 540, 2000, 8.4),
        Novel("Di Tengah Sejarah", 480, 1999, 7.8),
        Novel("Perjalanan Sejarah", 510, 2004, 8.3),
        Novel("Cerita dari Zaman Dulu", 460, 2007, 8.0),
        Novel("Runtuhnya Kerajaan", 530, 1996, 8.2),
        Novel("Menyusuri Zaman", 470, 1995, 8.4),
        Novel("Kisah Para Pejuang", 520, 2001, 8.5),
        Novel("Kehidupan di Masa Lalu", 460, 2012, 7.9),
        Novel("Perang yang Mengubah Dunia", 480, 1994, 8.6),
        Novel("Kekuatan Sejarah", 550, 2011, 8.3),
        Novel("Dunia yang Hilang", 510, 2009, 8.2),
        Novel("Pahlawan Abad Ini", 530, 2008, 8.7),
        Novel("Kisah Sejarah Dunia", 550, 2013, 8.4),
        Novel("Menghidupkan Sejarah", 420, 2015, 7.8),
        Novel("Membaca Zaman", 460, 2006, 8.1),
        Novel("Abad yang Lalu", 530, 1992, 8.3),
        Novel("Perjalanan dalam Sejarah", 500, 1990, 8.0),
        Novel("Menceritakan Kembali", 480, 2003, 7.7),
        Novel("Keajaiban Dunia", 550, 2017, 8.5),
        Novel("Langkah Sejarah", 470, 2018, 8.2),
        Novel("Menjadi Bagian dari Sejarah", 520, 2019, 8.6),
        Novel("Sejarah yang Menyentuh Hati", 540, 2010, 8.4),
        Novel("Raja-Raja dan Sejarah", 450, 2014, 7.9),
        Novel("Kisah Sukses di Abad Tua", 560, 2002, 8.1),
        Novel("Hikayat Sejarah Dunia", 500, 1998, 8.5),
        Novel("Jejak Perjalanan Sejarah", 480, 1991, 8.3),
        Novel("Legenda dan Kenangan", 470, 2000, 8.2),
        Novel("Sejarah yang Terungkap", 450, 1997, 8.4),
        Novel("Cerita Dari Zaman yang Lalu", 530, 1996, 8.5),
        Novel("Perjalanan Zaman", 490, 2005, 8.6),
        Novel("Kisah Perubahan Dunia", 550, 2004, 7.8),
        Novel("Zaman yang Berubah", 510, 2007, 8.1),
        Novel("Abad yang Mengubah Dunia", 560, 2012, 8.2),
        Novel("Pahlawan Sejati", 540, 1994, 8.4),
        Novel("Bukti Sejarah", 520, 2013, 7.9),
        Novel("Abad Kejayaan", 530, 2011, 8.3),
        Novel("Mengukir Sejarah", 500, 2008, 8.7),
        Novel("Dunia yang Berbeda", 480, 2002, 8.0),
        Novel("Cerita di Balik Sejarah", 510, 1993, 8.5),
        Novel("Perjalanan Luar Biasa", 460, 2001, 8.2),
        Novel("Peradaban yang Terlupakan", 520, 1995, 8.3),
        Novel("Meninggalkan Jejak", 540, 2010, 8.1),
        Novel("Badai Perubahan", 560, 1999, 8.4),
        Novel("Pahlawan Tanpa Nama", 500, 2016, 8.0),
        Novel("Perubahan di Dunia Lama", 470, 2019, 8.2),
        Novel("Titik Balik Sejarah", 530, 1998, 8.6),
        Novel("Dalam Bayang Sejarah", 510, 2015, 8.3),
        Novel("Membangun Dunia Baru", 550, 2006, 8.0),
        Novel("Di Tengah Pergolakan", 520, 2004, 8.2),
        Novel("Sejarah yang Menyakitkan", 540, 2014, 7.9),
        Novel("Kekuatan Sejarah", 470, 2017, 8.4),
        Novel("Menemukan Jejak", 490, 2009, 8.1),
        Novel("Tantangan Zaman", 500, 2008, 8.2),
        Novel("Jejak yang Menghilang", 460, 1992, 8.3),
        Novel("Cerita yang Terlupakan", 480, 1997, 7.8),
        Novel("Keajaiban Dunia Lama", 550, 2000, 8.6),
        Novel("Dunia yang Penuh Misteri", 510, 2018, 8.1),
        Novel("Sejarah yang Terulang", 530, 2003, 8.0),
        Novel("The Biology of Life", 450, 2011, 8.3),
        Novel("Genetics: A Tale of Evolution", 350, 2003, 8.5),
        Novel("In the Heart of Nature", 480, 2007, 7.8),
        Novel("The Cells Within", 500, 2015, 8.6),
        Novel("Molecules of Destiny", 550, 2009, 8.0),
        Novel("The Secrets of DNA", 450, 2012, 8.4),
        Novel("The Evolutionary Path", 420, 1999, 8.2),
        Novel("Survival of the Fittest", 480, 2000, 8.1),
        Novel("Biological Wonders", 500, 1995, 8.3),
        Novel("The Web of Life", 470, 2014, 8.5),
        Novel("The Anatomy of Growth", 450, 2010, 7.9),
        Novel("Symbiosis: The Hidden Story", 520, 2018, 8.7),
        Novel("Genes in Motion", 500, 2016, 8.2),
        Novel("The Microbial World", 480, 2002, 8.4),
        Novel("The Nature of Evolution", 550, 2004, 8.1),
        Novel("The Biology of Reproduction", 520, 1998, 8.5),
        Novel("The Secret Life of Plants", 470, 2006, 7.8),
        Novel("Life's Blueprint", 540, 2008, 8.6),
        Novel("From Cells to Ecosystems", 560, 2017, 8.0),
        Novel("Darwin’s Journey", 490, 1997, 8.3),
        Novel("The Ocean’s Pulse", 530, 2001, 8.2),
        Novel("The World of Microbes", 480, 1993, 8.0),
        Novel("Adaptation: A Tale of Survival", 500, 2012, 8.4),
        Novel("A Walk Through Evolution", 550, 2005, 8.1),
        Novel("The Diversity of Life", 510, 2013, 8.5),
        Novel("Cells in Motion", 470, 1999, 7.7),
        Novel("The Wonders of the Brain", 480, 2002, 8.0),
        Novel("The Mystery of Life’s Origins", 460, 2014, 8.2),
        Novel("Ecology: The Web of Nature", 530, 2007, 8.6),
        Novel("The Genetic Code", 490, 2011, 8.3),
        Novel("The Life of a Species", 520, 1995, 8.0),
        Novel("Nature's Blueprint", 510, 2015, 8.1),
        Novel("The Life Cycle", 500, 2008, 8.4),
        Novel("The Dynamics of Life", 540, 2000, 8.2),
        Novel("In the Garden of Evolution", 460, 2017, 8.5),
        Novel("The Survival Game", 470, 2016, 7.9),
        Novel("The Secret of Nature", 490, 2003, 8.3),
        Novel("In the Shadow of Life", 510, 2004, 8.1),
        Novel("From Genes to Organisms", 520, 2010, 8.6),
        Novel("The Brain and Behavior", 530, 2009, 8.2),
        Novel("Life in Balance", 540, 2012, 8.3),
        Novel("The Power of Evolution", 500, 2001, 8.5),
        Novel("The Cell's Journey", 480, 2006, 8.0),
        Novel("The Ecology of Life", 460, 2013, 7.8),
        Novel("The Biology of Love", 550, 2007, 8.4),
        Novel("From Molecules to Man", 510, 2005, 8.1),
        Novel("The Genetic Revolution", 470, 2014, 8.3),
        Novel("The Invisible World", 500, 1994, 8.2),
        Novel("Life Beyond Earth", 480, 2011, 7.9),
        Novel("Inside the Cell", 530, 2000, 8.5),
        Novel("The Evolutionary Force", 540, 1998, 8.0),
        Novel("Symbiosis and Survival", 510, 2016, 8.4),
        Novel("The Language of Life", 520, 2003, 8.2),
        Novel("Kabut Senja", 270, 2011, 8.0),
        Novel("Langkah Tak Terhenti", 340, 2018, 8.5),
        Novel("Batas Angkasa", 390, 2022, 9.0)
    ]


def display_novels(novels):
    for i, novel in enumerate(novels, start=1):
        print(f"{i}) {novel.name}, {novel.pages} halaman, {novel.year}, Rating: {novel.rating}")


def recursive_insertion_sort(novels, n=None, ascending=True):
    if n is None:
        n = len(novels)

    # Base case: if the array is fully sorted
    if n <= 1:
        return

    # Sort the first n-1 elements recursively
    recursive_insertion_sort(novels, n - 1, ascending)

    # Insert the nth element into the sorted part of the array
    key = novels[n - 1]
    j = n - 2

    # Loop from j down to 0 and shift elements
    while j >= 0 and ((ascending and novels[j].rating > key.rating) or (not ascending and novels[j].rating < key.rating)):
        novels[j + 1] = novels[j]
        j -= 1

    # Place the key element at the correct position
    novels[j + 1] = key



def find_max_iterative(novels):
    return max(novels, key=lambda novel: novel.rating)


def find_min_iterative(novels):
    return min(novels, key=lambda novel: novel.rating)


def main():
    novels = init_novels()

    print("Data Novel (Awal):")
    display_novels(novels)

    # Sort novels in ascending order
    recursive_insertion_sort(novels, ascending=True)
    print("\nData Novel Setelah Disortir (Ascending):")
    display_novels(novels)

    # Sort novels in descending order
    recursive_insertion_sort(novels, ascending=False)
    print("\nData Novel Setelah Disortir (Descending):")
    display_novels(novels)

    # Find the novel with the highest rating
    max_novel = find_max_iterative(novels)
    print(f"\nNovel dengan Rating Tertinggi: {max_novel.name} dengan rating {max_novel.rating}")

    # Find the novel with the lowest rating
    min_novel = find_min_iterative(novels)
    print(f"Novel dengan Rating Terendah: {min_novel.name} dengan rating {min_novel.rating}")


if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import random
import time
plt.style.use('ggplot')
class Novel:
    def __init__(self, title, pages, year, rating):
        self.title = title
        self.pages = pages
        self.year = year
        self.rating = rating

def recursive_insertion_sort(novels, n=None, ascending=True):
    if n is None:
        n = len(novels)

    # Base case: if the array is fully sorted
    if n <= 1:
        return

    # Sort the first n-1 elements recursively
    recursive_insertion_sort(novels, n - 1, ascending)

    # Insert the nth element into the sorted part of the array
    key = novels[n - 1]
    j = n - 2

    # Loop from j down to 0 and shift elements
    while j >= 0 and ((ascending and novels[j].rating > key.rating) or (not ascending and novels[j].rating < key.rating)):
        novels[j + 1] = novels[j]
        j -= 1

    # Place the key element at the correct position
    novels[j + 1] = key

def insertion_sort_iterative(novels, ascending=True):
    n = len(novels)
    for i in range(1, n):
        key = novels[i]
        j = i - 1
        while j >= 0 and ((ascending and novels[j].rating > key.rating) or (not ascending and novels[j].rating < key.rating)):
            novels[j + 1] = novels[j]
            j -= 1
        novels[j + 1] = key

# Generate dummy data
novels = [Novel(f"Novel {i+1}", random.randint(100, 1000), random.randint(1900, 2024), round(random.uniform(1, 10), 1)) for i in range(350)]

# Prepare data for timing
sizes = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350]
recursive_times = []
iterative_times = []

for size in sizes:
    subset = novels[:size]

    # Measure recursive execution time
    novels_copy = subset.copy()
    start_time = time.time()
    recursive_insertion_sort(novels_copy)
    recursive_times.append(time.time() - start_time)

    # Measure iterative execution time
    novels_copy = subset.copy()
    start_time = time.time()
    insertion_sort_iterative(novels_copy)
    iterative_times.append(time.time() - start_time)

# Plot the execution times
plt.figure(figsize=(10, 6))
plt.plot(sizes, recursive_times, label="Recursive Insertion Sort", marker="o", linestyle="--", color="red")
plt.plot(sizes, iterative_times, label="Iterative Insertion Sort", marker="x", linestyle="-", color="blue")
plt.title("Execution Time Comparison: Recursive vs Iterative Insertion Sort")
plt.xlabel("Number of Novels")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()

# Print execution time arrays
print("\nExecution Times:")
print(f"Recursive: {recursive_times}")
print(f"Iterative: {iterative_times}")
```
